````python
from bs4 import BeautifulSoup
````

````python
soup = BeautifulSoup(req.text, "lxml")
````

````python
soup
````

````python
soup.find_all('span')
````

````python
soup.find_all('span', attrs={'class': 'weather-type-image'})
````

````python
daily = soup.find('ul', attrs={'class': 'daily'})
````

````python
daily
````

````python
daily.find_all('li')
````

````python
data = []
for li in daily.find_all('li'):
    day = {}
    day_span = li.find('span', attrs={'class': 'day-name'})
    day['name'] = day_span.attrs['aria-label']
    rain_span = li.find('span', attrs={'class': 'weather-type-image'})
    day['rain_type'] = rain_span.attrs['title']
    max_temp_c = li.select('span.max-temp span.units-value')[0]
#     print(max_temp_c)
    print(max_temp_c.attrs)
#     print(max_temp_c.text)
    print(list(max_temp_c.strings))
    data.append(day)
print(data)
````
