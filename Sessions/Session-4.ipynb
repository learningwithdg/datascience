```python
data = {
    'pokemons': []
}
with open('../Data/Pokemon.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile, fieldnames=fieldnames)
    next(reader)
    for line in reader:
        data['pokemons'].append(line)

with open('../Data/Pokemon.json', 'w') as jsonfile:
    json.dump(data, jsonfile)
```

## Sqlite

```python
import sqlite3
```

```python
conn = sqlite3.connect('../Data/db.sqlite3')
```

```python
conn
```

```python
cursor = conn.cursor()
```

```python
cursor
```

```python
sql = """CREATE TABLE pokemons(id INTEGER primary key, attack INTEGER, defense INTEGER,
                               generation INTEGER, hp INTEGER, legendary BOOLEAN, 
                               name TEXT, sp_atk INTEGER, sp_def INTEGER, speed INTEGER,
                               total INTEGER, type1 TEXT, type2 TEXT)"""
```

```python
cursor.execute(sql)
```

```python
conn.commit()
```

```python
sql = """INSERT INTO pokemons(id, attack, defense, generation, hp, legendary, name, sp_atk,
                              sp_def, speed, total, type1, type2) 
                              VALUES(1, 49, 49, 1, 45, 0, 'Bulbasaur', 65, 65, 45, 318,
                                     'Grass', 'Poison')"""
```

```python
cursor.execute(sql)
```

```python
conn.commit()
```

```python
sql = """INSERT INTO pokemons(id, attack, defense, generation, hp, legendary, name, sp_atk,
                              sp_def, speed, total, type1, type2) 
                              VALUES(?,?,?,?,?,?,?,?,?,?,?,?,?)"""
```

```python
counter = 1
with open('../Data/Pokemon.json', 'r') as _jsonfile:
    _json = json.load(_jsonfile)
    for each in _json['pokemons']:
        _leg = 0 if each['Legendary'] == 'False' else 1
        values = (counter, int(each['Attack']), int(each['Defense']),
                  int(each['Generation']), int(each['HP']), _leg, 
                  each['Name'], int(each['Sp. Atk']), int(each['Sp. Def']),
                  int(each['Speed']), int(each['Total']), each['Type 1'], each['Type 2'])
        print(values)
        cursor.execute(sql, values)
        counter += 1
conn.commit()
```
