**Let's plot sin**

$$y = sin(x) \text{ where }x=\pi\to-\pi$$

```python
x = np.linspace(-np.pi, np.pi, num=100)
x
```

```python
y = np.sin(x)
y
```

**Let's plot sin and cos side by side**

$$y = cos(x) \text{ where }x = \pi\to-\pi$$

```python
plt.plot(x, np.cos(x))
```

*both in single plot*

```python
plt.plot(x, np.sin(x), x, np.cos(x))
```

*Matplotlib automatically uses different color for different lines*

*Or use consecutive plot command and show final plot*

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

*Above plots are displayed using matplotlib defaults, such as line color, label font, line width and so on.*

```python
# for example changing line color and line width
plt.plot(x, np.sin(x), color='red', linewidth=2.0)
plt.plot(x, np.cos(x), color='blue', linewidth=2.0)
plt.show()
```

**Changing limits**

```python
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin, color='red', linewidth=2.0)
plt.plot(x, y_cos, color='blue', linewidth=2.0)

# limiting plot values
plt.xlim(x.min()*1.2, x.max()*1.2)
plt.ylim(y_cos.min()*1.2, y_cos.max()*1.2)

plt.show()
```

*Above figure shows the whole signal, without clipping*

*__Note__: In above figures you will not see this(clipping) behavior if you are using latest matplotlib (v2 and above). But it is wise to know how to change the limits.*

**Setting up ticks**

```python
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin, color='red', linewidth=2.0)
plt.plot(x, y_cos, color='blue', linewidth=2.0)

# limiting plot values
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(y_cos.min()*1.1, y_cos.max()*1.1)

# ticks
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1])

plt.show()
```

*We can pass labels for the ticks, eg 2nd list on __plt.xticks__ is list of labels in __latex__*

**Adding legend**

```python
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin, color='red', linewidth=2.0, label='sine')
plt.plot(x, y_cos, color='blue', linewidth=2.0, label='cosine')

# limiting plot values
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(y_cos.min()*1.1, y_cos.max()*1.1)

# ticks
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1])

# show labels with legend on specific location
plt.legend(loc="upper left", frameon=False)

plt.show()
```

### Figure

Figure is a window, where we create plots and subplots.

*Figure constitute plots (either single or multiple), legends, boundry, title etc.*

```python
plt.figure(1)
plt.plot(np.linspace(0, 99))
plt.show()
```

### Subplots

Subplots are plots that are created inside figure

```python
plt.figure(1)
plt.subplot(211) # or plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.subplot(212) # or plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.show()
```

*plt.subplot(number_of_rows, number_of_columns, number_of_the_plot)*

Where above

*first two arguments( or numbers in above case ) plt.subplot(21)* denotes that this figure has two rows and single column. while last argument ( or number in above case ) __1__ and __2__ in __211__ and __212__ respectively says first subplot must be drawn in 1st axes ( an area inside figure where graphs, label etc are drawn ) and second subplot must be drawn in 2nd axes.

**For example if we have grid of 4 subplots**

```python
plt.figure(1)
# draw in first subplot ie. top-left one
plt.subplot(221) # or plt.subplot(2, 2, 1)
plt.plot(x, y_sin)

# draw in last subplot ie. bottom-right one
plt.subplot(224) # or plt.subplot(2, 2, 4)
plt.plot(x, y_cos)
plt.show()
```

## Axes

### Spines

```python
x_range = np.linspace(-np.pi, np.pi, num=256, endpoint=True)
sin_y = np.sin(x_range)
cos_y = np.cos(x_range)

plt.plot(x_range, sin_y, color='red', linewidth=2.0)
plt.plot(x_range, cos_y, color='blue', linewidth=2.0)

# limiting plot values
plt.xlim(x_range.min()*1.1, x_range.max()*1.1)
plt.ylim(cos_y.min()*1.1, cos_y.max()*1.1)

# ticks
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1])

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))


plt.show()
```
